```python
import requests
from bs4 import BeautifulSoup

def scrape_news_article(url):
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract article text
        article_text = ''
        article_text_tag = soup.find('div', class_='css-158dogj evys1bk0')
        if article_text_tag:
            article_text = article_text_tag.get_text(separator='\n', strip=True)

        # Extract classification label
        classification_label = ''
        classification_label_tag = soup.find('a', class_='css-1xf86vz')
        if classification_label_tag:
            classification_label = classification_label_tag.text.strip()

        # Extract URL
        article_url = url

        return article_text, classification_label, article_url
    else:
        return None, None, None  # Handle invalid URLs or errors

def scrape_news_articles(base_url, sections):
    scraped_data = []

    for section in sections:
        section_url = f'{base_url}{section}'
        response = requests.get(section_url)

        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')

            # Find all article links
            articles = soup.find_all('article')
            for article in articles:
              article_links = soup.find_all('a')
              for article_link in article_links:
                  article_url = ""
                  if "https://www.nytimes.com" in str(article_link):
                    article_url = article_link['href']
                  else:
                    article_url = base_url+article_link['href'][1:]

                  article_text, classification_label, article_url = scrape_news_article(article_url)

                  if article_text:
                      scraped_data.append({'article_text': article_text, 'classification_label': classification_label, 'url': article_url})
                      print(scraped_data)

    return scraped_data

if __name__ == '__main__':
    base_url = 'https://www.nytimes.com/section/'
    sections = ['world', 'US', 'business', 'technology', 'health', 'science', 'arts', 'sports']

    scraped_data = scrape_news_articles(base_url, sections)

    # Process or store the scraped data as needed
    print(scraped_data)
```
